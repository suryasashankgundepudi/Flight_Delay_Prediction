### Defining the list with the values of the airports that are under observation
1. This list comes from the problem statement

```python
airport_list = ['ATL', 'CLT', 'DEN', 'DFW', 'EWR', 'IAH', 'JFK', 'LAS', 
                'LAX', 'MCO', 'MIA', 'ORD', 'PHX', 'SEA', 'SFO']
```

### Getting the overall dataset for all the months of 2016
1. First we define a path to the file and since it only changes with the integer, which always increments by one, we just add the string of that integer
2. Then we filter out the file where the `Origin` and the `Destination` of the Flight is in the airport list as we defined above.
3. We append each dataframe to a larger dataframe which we use for later purposes

```python
# Creating the main overall dataframe
df_main1 = pd.DataFrame()

# Starting the `for` loop which helps us change the file name
for i in range(1, 13):
    # The path changes as the value of `i` changes
    path = r"Data\2016\On_Time_On_Time_Performance_2016_"+str(i)+'/'+"On_Time_On_Time_Performance_2016_"+str(i)+'.csv'
    df = pd.read_csv(path)
    
    #Filtering the dataframe based on the airport list
    df1 = df[df['Origin'].isin(airport_list)]
    df2 = df1[df1['Dest'].isin(airport_list)]
    # Creating a temporary dataframe to append to this new dataframe
    if i == 1:
        temp_df = pd.DataFrame()
        df_main1 = temp_df.append(df2,ignore_index = True)
    else:
        df_main1 = df_main1.append(df2,ignore_index = True)
```

```python
# Looking at the shape of the dataframe
df_main1.shape
```

#### Looking at one of the 2017 datasets
1. Let us now look at the dataset from one of the 2017 files

```python
df_main2 = pd.DataFrame()
for i in range(1, 13):
    path = r"Data\2016\On_Time_On_Time_Performance_2016_"+str(i)+'/'+"On_Time_On_Time_Performance_2016_"+str(i)+'.csv'
    df = pd.read_csv(path)
    df1 = df[df['Origin'].isin(airport_list)]
    df2 = df1[df1['Dest'].isin(airport_list)]
    if i == 1:
        temp_df = pd.DataFrame()
        df_main2 = temp_df.append(df2,ignore_index = True)
    else:
        df_main2 = df_main2.append(df2,ignore_index = True)
```

### Creating the final dataframe
1. Appending both of the 2016 and 2017 datasets

```python
final_df = df_main1.append(df_main2,ignore_index = True)
```

### Filtering out
These are the variables that we will be looking at in the future

```python
cols = ['ArrDel15', 'ArrDelayMinutes', 'ArrTime', 'CRSArrTime', 'CRSDepTime', 'DayofMonth', 'DepDel15', 'DepDelayMinutes', 
        'DepTime', 'DestAirportID', 'FlightDate', 'Month', 'Origin', 'OriginAirportID', 'Quarter', 'Year']
```
